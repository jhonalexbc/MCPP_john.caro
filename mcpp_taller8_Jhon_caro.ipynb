```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re
import pandas as pd
import numpy as np
```

```python
base_url = "http://listas.20minutos.es/lista/500-mejores-peliculas-de-la-historia-segun-la-revista-empire-297804/"
```

```python
index_html = urlopen(base_url)
```

```python
index = BeautifulSoup(index_html, "lxml")
```

```python
info1 = index.find_all ("h3")
```

```python
mstring = "".join(map(str,info1))
```

```python
title = re.findall("</strong>(.*?)</h3>",mstring)
```

```python
ranking = re.findall("<strong>([^.].*?)</strong>",mstring)
```

```python
info2 = index.find_all("div", {"class": "descripcion_elemento"})
```

```python
mstring2 = "".join(map(str,info2))
```

```python
fecha =re.findall("[0-9]+\r|[0-9]+<",mstring2)
fechas = []
for i in range(len(fecha)):
    fecha2 = str(fecha[i])
    fechas.append(fecha2.rstrip("\r"))
fecha = fechas
```

```python
info3 = index.find_all("div", {"class": "descripcion_elemento_larga"})
```

```python
mstring3 = "".join(map(str,info3))
```
