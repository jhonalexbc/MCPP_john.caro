# 3.2

```python
def limacon(z,theta):
    r = z + np.cos(theta)
    x = r * np.cos(theta)
    y = r * np.sin(theta)
    return r,x,y
```

```python
z = np.array([1])
theta = np.linspace(-4.5,4.5,2000)
```

```python
limacons = limacon(z,theta)
```

```python
xs1 = limacons[1]
ys1 = limacons[2]
```

```python
plt.plot(xs1,ys1,"g",label="r0 = 1")
plt.legend()
plt.savefig("limaco con ro = 1.jpg")
```

```python
z = np.array([0.8])
theta = np.linspace(-4.5,4.5,2000)
limacons = limacon(z,theta)
xs2 = limacons[1]
ys2 = limacons[2]
```

```python
plt.plot(xs2,ys2,"r",label="r0 = 0.8")
plt.legend()
plt.savefig("limaco con ro = 0,8.jpg")
```

```python
z = np.array([1.2])
theta = np.linspace(-4.5,4.5,2000)
limacons = limacon(z,theta)
xs3 = limacons[1]
ys3 = limacons[2]
```

```python
plt.plot(xs3,ys3,label="r0 = 1.2")
plt.legend()
plt.savefig("limaco con ro = 1,2.jpg")
```

```python
plt.plot(xs1,ys1,"g",label="r0 = 1", linewidth=3)
plt.plot(xs2,ys2,"r",label="r0 = 0.8")
plt.plot(xs3,ys3,label="r0 = 1.2")
plt.legend()
plt.savefig("limacos.jpg")
```
